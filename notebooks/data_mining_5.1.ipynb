Ejemplo básico de una serie temporal que iremos construyendo paso a paso

```
# librerias necesarias, las básicas de seimpre

import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates    # formato data time
```

```
# Luego importaremos los datos pero de momento creamos una pequeña lista
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]
```

```
y = [0, 1, 3, 4, 6, 5, 7]
```

y el ejemplo más sencillo donde x son las fechas e y las observaciones tendríamos esta gráfica

de momento no el más bonito, pero plot x e y, directo

```
plt.plot_date (dates, y)
```

por defecto nos lo pone sin líneas pero se puede unir indicando linestyle='solid'

```

plt.plot_date(dates, y, linestyle='solid')
```

**Adelante**, ya podéis empezar a ver algunos de vuestros datos y cómo evolucionan

Forzamos a más pequeño y con autofmt rota

```
plt.figure(figsize=(6,6))
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()
```

```
# importamos y restos de pasos con pandas que conocéis de sobra y ejemplo básico de consttrucción de gráfica con matplot
data = pd.read_csv('DataSimple.csv')

#creo una nueva columna
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.figure(figsize=(12,6), dpi=100)

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()

```
